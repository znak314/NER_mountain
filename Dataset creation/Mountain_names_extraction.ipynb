```python
df = pd.read_csv("Mountain.csv")
df.head()
```

## Choose 200 random mountains

```python
df = df["Mountain"]

random_200_mountains = df.sample(n=200, random_state=42)
```

```python
mountain_list = random_200_mountains.tolist()

with open('mountains_names.txt', 'w') as f:
    for mountain in mountain_list:
        f.write(mountain + '\n')
```
